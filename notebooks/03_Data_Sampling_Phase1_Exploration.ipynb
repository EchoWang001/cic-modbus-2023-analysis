```python
print("=" * 60)
print("Deep Diagnosis: External PCAP vs CSV Timestamps")
print("=" * 60)

#  1. Read External PCAP timestamps
from scapy.all import rdpcap, TCP
from datetime import datetime, timedelta
import pandas as pd

external_pcap = r"C:\Users\Echo\Desktop\Modbus Dataset\Modbus Dataset\attack\external\external-attacker\external-attacker-network-capture\veth665f3cf-0.pcap"

print("\n1. Reading PCAP timestamps (first 10 Modbus packets):")
packets = rdpcap(external_pcap, count=1000)

pcap_times = []
for pkt in packets:
    if TCP in pkt and (pkt[TCP].sport == 502 or pkt[TCP].dport == 502):
        try:
            payload = bytes(pkt[TCP].payload)
            if len(payload) >= 8:
                ts_raw = datetime.fromtimestamp(float(pkt.time))
                ts_plus3 = ts_raw + timedelta(hours=3)
                pcap_times.append({
                    'raw': ts_raw,
                    'plus3': ts_plus3
                })
                if len(pcap_times) >= 10:
                    break
        except:
            pass

print("\nPCAP timestamps (first 10):")
for i, t in enumerate(pcap_times[:10], 1):
    print(f"  {i}. Raw: {t['raw']}")
    print(f"      +3h: {t['plus3']}")

#  2. Read CSV label timestamps
import glob
external_csvs = []
for root, dirs, files in os.walk(config.EXTERNAL_LABELS_DIR):
    external_csvs.extend([os.path.join(root, f) for f in files if f.endswith('.csv')])

print(f"\n2. Reading CSV label timestamps (from {len(external_csvs)} files):")

all_csv_times = []
for csv_file in external_csvs:
    df = pd.read_csv(csv_file)
    for ts in df['Timestamp']:
        try:
            parsed = pd.to_datetime(ts)
            all_csv_times.append(parsed)
        except:
            pass

#  Display CSV time range
if all_csv_times:
    all_csv_times_sorted = sorted(all_csv_times)
    print(f"\nCSV timestamp range:")
    print(f"  Earliest: {all_csv_times_sorted[0]}")
    print(f"  Latest: {all_csv_times_sorted[-1]}")
    print(f"  Total: {len(all_csv_times)} records")
    
    print(f"\nFirst 10 CSV timestamps:")
    for i, t in enumerate(all_csv_times_sorted[:10], 1):
        print(f"  {i}. {t}")

#  3. Calculate time difference
print("\n3. Time difference analysis:")
if pcap_times and all_csv_times:
    pcap_first = pcap_times[0]['raw']
    pcap_first_plus3 = pcap_times[0]['plus3']
    csv_first = all_csv_times_sorted[0]
    
    diff_raw = abs((pcap_first - csv_first).total_seconds())
    diff_plus3 = abs((pcap_first_plus3 - csv_first).total_seconds())
    
    print(f"PCAP first packet (raw): {pcap_first}")
    print(f"PCAP first packet (+3h): {pcap_first_plus3}")
    print(f"CSV first label:         {csv_first}")
    print(f"\nTime diff (raw): {diff_raw:.1f} sec = {diff_raw/3600:.1f} hours")
    print(f"Time diff (+3h): {diff_plus3:.1f} sec = {diff_plus3/3600:.1f} hours")
    
    #  Try different timezone conversions
    print("\n4. Trying different timezone conversions:")
    for hours in [-3, -2, -1, 0, 1, 2, 3, 4, 5, 6]:
        adjusted = pcap_first + timedelta(hours=hours)
        diff = abs((adjusted - csv_first).total_seconds())
        marker = "  ✓✓✓" if diff < 60 else ""
        print(f"  {hours:+2d}h: {adjusted} → diff {diff:.1f}s {marker}")

#  5. Check if PCAP and CSV dates match
print("\n5. Date matching check:")
if pcap_times and all_csv_times:
    pcap_dates = set([t['raw'].date() for t in pcap_times])
    csv_dates = set([t.date() for t in all_csv_times_sorted])
    
    print(f"PCAP dates: {sorted(pcap_dates)}")
    print(f"CSV dates:  {sorted(csv_dates)}")
    
    common_dates = pcap_dates & csv_dates
    if common_dates:
        print(f"✓ Overlapping dates found: {sorted(common_dates)}")
    else:
        print(f"✗ No overlapping dates! PCAP and CSV may be from different time periods")
```

```python
print("=" * 60)
print("Precise Comparison: 2023-02-01 PCAP vs CSV")
print("=" * 60)

from datetime import date

#  1. Extract Feb 1st PCAP timestamps (all Modbus packets)
target_date = date(2023, 2, 1)

print("\n1. Extracting all Feb 1st PCAP Modbus packets...")
external_pcap = r"C:\Users\Echo\Desktop\Modbus Dataset\Modbus Dataset\attack\external\external-attacker\external-attacker-network-capture\veth665f3cf-0.pcap"

packets = rdpcap(external_pcap)
print(f"Total packets: {len(packets)}")

pcap_times_0201 = []
for pkt in packets:
    if TCP in pkt and (pkt[TCP].sport == 502 or pkt[TCP].dport == 502):
        try:
            payload = bytes(pkt[TCP].payload)
            if len(payload) >= 8:
                ts_raw = datetime.fromtimestamp(float(pkt.time))
                
                #  Only keep Feb 1st data
                if ts_raw.date() == target_date:
                    ts_plus3 = ts_raw + timedelta(hours=3)
                    pcap_times_0201.append({
                        'timestamp': ts_raw,
                        'timestamp_utc': ts_plus3,
                        'txid': int.from_bytes(payload[0:2], byteorder='big'),
                        'dst_ip': pkt['IP'].dst
                    })
        except:
            pass

print(f"✓ Feb 1st Modbus packets: {len(pcap_times_0201)}")

if len(pcap_times_0201) > 0:
    times_sorted = sorted([t['timestamp'] for t in pcap_times_0201])
    print(f"\nPCAP time range (raw):")
    print(f"  Earliest: {times_sorted[0]}")
    print(f"  Latest: {times_sorted[-1]}")
    
    times_sorted_utc = sorted([t['timestamp_utc'] for t in pcap_times_0201])
    print(f"\nPCAP time range (+3h):")
    print(f"  Earliest: {times_sorted_utc[0]}")
    print(f"  Latest: {times_sorted_utc[-1]}")

#  2. Extract Feb 1st CSV labels
print("\n2. Extracting Feb 1st CSV labels...")

csv_0201 = []
for csv_file in external_csvs:
    df = pd.read_csv(csv_file)
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
    df = df.dropna(subset=['Timestamp'])
    
    #  Only keep Feb 1st data
    df_0201 = df[df['Timestamp'].dt.date == target_date]
    
    if len(df_0201) > 0:
        print(f"  File: {os.path.basename(csv_file)} - {len(df_0201)} records")
        csv_0201.append(df_0201)

if csv_0201:
    df_csv_0201 = pd.concat(csv_0201, ignore_index=True)
    print(f"\n✓ Feb 1st CSV labels: {len(df_csv_0201)} records")
    
    print(f"\nCSV time range:")
    print(f"  Earliest: {df_csv_0201['Timestamp'].min()}")
    print(f"  Latest: {df_csv_0201['Timestamp'].max()}")
    
    #  3. Precise matching test
    print("\n" + "=" * 60)
    print("3. Precise matching test (Feb 1st data)")
    print("=" * 60)
    
    #  Create PCAP DataFrame
    df_pcap_0201 = pd.DataFrame(pcap_times_0201)
    
    #  Test different time window and timezone combinations
    print("\nTesting different configurations:")
    
    for tz_hours in [-3, 0, 3]:
        for time_window in [1, 5, 10, 30]:
            df_pcap_test = df_pcap_0201.copy()
            
            if tz_hours == 0:
                df_pcap_test['ts'] = df_pcap_test['timestamp']
            else:
                df_pcap_test['ts'] = df_pcap_test['timestamp'] + timedelta(hours=tz_hours)
            
            matched = 0
            for _, label in df_csv_0201.iterrows():
                time_mask = abs((df_pcap_test['ts'] - label['Timestamp']).dt.total_seconds()) <= time_window
                if time_mask.sum() > 0:
                    matched += 1
            
            match_rate = matched / len(df_csv_0201) * 100
            marker = "  ✓✓✓" if match_rate > 10 else ""
            print(f"  TZ{tz_hours:+2d}h, window±{time_window:2d}s: {matched:5d}/{len(df_csv_0201)} = {match_rate:5.1f}%{marker}")
    
    #  4. Display specific matching examples
    print("\n" + "=" * 60)
    print("4. View specific timestamp comparison (first 5 CSV records)")
    print("=" * 60)
    
    for i, (_, label) in enumerate(df_csv_0201.head(5).iterrows(), 1):
        print(f"\nCSV label #{i}:")
        print(f"  Time: {label['Timestamp']}")
        print(f"  Attack: {label['Attack']}")
        
        #  Find closest PCAP packet (raw time)
        df_pcap_test = df_pcap_0201.copy()
        df_pcap_test['time_diff'] = abs((df_pcap_test['timestamp'] - label['Timestamp']).dt.total_seconds())
        closest = df_pcap_test.nsmallest(3, 'time_diff')
        
        print(f"  Closest PCAP packets (raw time):")
        for j, (_, p) in enumerate(closest.iterrows(), 1):
            print(f"    {j}. {p['timestamp']} (diff {p['time_diff']:.1f}s)")
        
        #  Find closest PCAP packet (+3h time)
        df_pcap_test['time_diff'] = abs((df_pcap_test['timestamp_utc'] - label['Timestamp']).dt.total_seconds())
        closest = df_pcap_test.nsmallest(3, 'time_diff')
        
        print(f"  Closest PCAP packets (+3h time):")
        for j, (_, p) in enumerate(closest.iterrows(), 1):
            print(f"    {j}. {p['timestamp_utc']} (diff {p['time_diff']:.1f}s)")

else:
    print("✗ No Feb 1st CSV labels found")
```
