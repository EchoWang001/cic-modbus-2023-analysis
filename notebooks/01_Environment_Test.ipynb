```python
#  Cell 2: Test Read PCAP
from scapy.all import rdpcap, TCP

#  Setup Dataset Path
dataset_path = r"C:\Users\Echo\Desktop\Modbus Dataset\Modbus Dataset"
test_pcap = dataset_path + r"\benign\network-wide-pcap-capture\network-wide\network-wide-normal-14.pcap"

print(f"Reading: {test_pcap}")

#  Read PCAP
packets = rdpcap(test_pcap)
print(f"✓ Successfully read {len(packets):,} packets")

#  Statistics Modbus Packets
modbus_packets = [p for p in packets if TCP in p and (p[TCP].sport == 502 or p[TCP].dport == 502)]
print(f"✓ Modbus packets: {len(modbus_packets):,}")
```
